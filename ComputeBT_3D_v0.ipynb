```python
import xarray as xr
import numpy as np
import datetime as dt
import os
import glob
import utm
import pandas as pd
import monthdelta
import matplotlib.pyplot as plt
from scipy.interpolate import RectBivariateSpline
import copy
```

```python
def read_era_file(datetime, path_era5='/Users/jhayron/Data/ERA5/'):
    year = str(datetime.year)
    month = str(datetime.month)
    path_file = f'{path_era5}{year}/ERA5_{year}_{month.zfill(2)}.nc'
    return xr.open_dataset(path_file)

def erase_points_below_surface(dataset,
    path_geop_surface = '/Users/jhayron/Data/ERA5/geo_1279l4_0.1x0.1.grib2_v4_unpack.nc'):
    dataset_surface = xr.open_dataset('/Users/jhayron/Data/ERA5/geo_1279l4_0.1x0.1.grib2_v4_unpack.nc')
    interp_spline = RectBivariateSpline(dataset_surface.latitude.values[::-1],
                                        dataset_surface.longitude.values,
                                        dataset_surface.z.values[0][::-1])
    z2 = interp_spline(dataset.latitude.values[::-1], dataset.longitude.values+360)[::-1]
    dataset = dataset.where(dataset.z-z2>100)
    return dataset

def interpolate_field_2_grid(dataarray,lat_final,lon_final):
    interp_spline = RectBivariateSpline(dataarray.latitude.values[::-1],
                                        dataarray.longitude.values,
                                        dataarray.values[::-1])

    array2 = interp_spline(lat_final[::-1], lon_final)[::-1]
    array2 = xr.DataArray(array2,
                          dims=dataarray.dims,
                          coords = {"longitude": lons_start,"latitude": lats_start},
                          attrs = dataarray.attrs)
    return array2
```

```python
def compute_location(lat0,lon0,plev0,u,v,w,delta_t,BT=True):
    '''
    The ‘zero acceleration’ solution to the differential trajectory equation
        X(t_1) = X(t_0)+(delta_t)*V(t_0)


    u,v:     m/s
    omega:   Pascals/s
    delta_t: horas
    '''
    
    # Forward Trajectories
    dx = u*delta_t*60*60       # meters
    dy = v*delta_t*60*60       # meters
    dz = (w/100)*delta_t*60*60 # hPa
    
    plev = plev0+(-dz if BT else dz)
    r_earth = 6378000
    if BT:
        # Transformando metros a grados
        new_latitude  = lat0 - (dy / r_earth) * (180 / np.pi)
        new_longitude = lon0 - (dx / r_earth) * (180 / np.pi) \
                                                / np.cos(lat0 * np.pi/180)
    else:
        new_latitude  = lat0 + (dy / r_earth) * (180 / np.pi)
        new_longitude = lon0 + (dx / r_earth) * (180 / np.pi) \
                                                / np.cos(lat0 * np.pi/180)
#     if plev>1000:
#         plev = 1000
    return new_latitude, new_longitude, plev

def interpolate_field_2_grid_all_levels(dataarray,lat_final_2d,lon_final_2d,lats_start,lons_start):
    array_new = np.zeros([len(dataarray.level),len(lat_final_2d[:,0]),len(lon_final_2d[0,:])])
    for leveli in range(len(dataarray.level)):
        interp_spline = RectBivariateSpline(dataarray.latitude.values[::-1],
                                                dataarray.longitude.values,
                                                dataarray.values[leveli][::-1])
        array_new[leveli] = np.reshape(interp_spline(lat_final_2d.values.ravel(),lon_final_2d.values.ravel(),grid=False),
           [len(lat_final_2d),len(lon_final_2d)])
    
    array_new = xr.DataArray(array_new,
                          dims=dataarray.dims,
                          coords = {"level": dataarray.level,"longitude": lons_start,"latitude": lats_start},
                          attrs = dataarray.attrs)
    return array_new
```

```python
dataset = read_era_file(dt.datetime(2020,6,1))
dataset_atm = erase_points_below_surface(dataset)
```

```python
lats_start = np.arange(0,10,0.1)[::-1]
lons_start = np.arange(-80,-70,0.1)
lons_start_2d,lats_start2d = np.meshgrid(lons_start,lats_start)
```

```python
init_time = dataset.time[1]
delta_t = 2 #hora
days_back = 5
hours_back = days_back*24/delta_t
```

```python
# 1. Initial conditions

dataset_temp = dataset_atm.sel(time = init_time,drop=True)
ds_init = dataset_temp.isel(level = dataset_temp.q.argmax('level'))

z_interpolated = interpolate_field_2_grid(ds_init.z,lats_start,lons_start)
u_interpolated = interpolate_field_2_grid(ds_init.u,lats_start,lons_start)
v_interpolated = interpolate_field_2_grid(ds_init.v,lats_start,lons_start)
w_interpolated = interpolate_field_2_grid(ds_init.w,lats_start,lons_start)
q_interpolated = interpolate_field_2_grid(ds_init.q,lats_start,lons_start)
r_interpolated = interpolate_field_2_grid(ds_init.r,lats_start,lons_start)
p_interpolated = interpolate_field_2_grid(ds_init.level,lats_start,lons_start)

# 2. Create storage variables

z_traj = copy.deepcopy(z_interpolated)
u_traj = copy.deepcopy(u_interpolated)
v_traj = copy.deepcopy(v_interpolated)
w_traj = copy.deepcopy(w_interpolated)
q_traj = copy.deepcopy(q_interpolated)
r_traj = copy.deepcopy(r_interpolated)
p_traj = copy.deepcopy(p_interpolated)

time_traj = []

lat_new2d = copy.deepcopy(lats_start2d)
lon_new2d = copy.deepcopy(lons_start_2d)
p_new2d = copy.deepcopy(p_interpolated)

time_new = copy.deepcopy(init_time.values)
time_traj.append(time_new)

lat_traj = []
lon_traj = []
lat_traj.append(lat_new2d)
lon_traj.append(lon_new2d)
print(dt.datetime.now())
for ti in range(int(hours_back)):
# for ti in range(int(4)):
#     print(time_new)
    # 3. Compute past location
    lat_new2d,lon_new2d,p_new2d = compute_location(lat_new2d,
                                                   lon_new2d,
                                                   p_new2d,
                                                   u_interpolated,
                                                   v_interpolated,
                                                   w_interpolated,
                                                   delta_t)
    lat_traj.append(lat_new2d)
    lon_traj.append(lon_new2d)
#     aaaa
    # 4. Collect z, u, v, w, q, r, p at new location

    time_new = time_new - np.timedelta64(delta_t,'h')
    time_traj.append(time_new)
    ds_new = dataset.isel(time = abs(dataset.time - time_new).argmin('time'),drop=True)

    # Initially I have to get the indices (levels) from which I want to take the info:
    # interpolate all fields at 8 levels to new lats and lons, get level at which pressure interpolated
    # is nearest to new pressure and extract the other variables.

    # Build a dummy pressure array

    p_aux = np.zeros_like(ds_new.z)
    p_aux[0] = 500
    p_aux[1] = 600
    p_aux[2] = 700
    p_aux[3] = 800
    p_aux[4] = 850
    p_aux[5] = 900
    p_aux[6] = 950
    p_aux[7] = 1000

    p_aux = xr.DataArray(p_aux,
                        dims = ds_new.z.dims,
                        coords = ds_new.z.coords,
                        attrs = ds_new.level.attrs)

    # Interpolate everything to the new points in the trajectories

    p_aux_interpolated = interpolate_field_2_grid_all_levels(p_aux,lat_new2d,lon_new2d,lats_start,lons_start)
    z_interpolated = interpolate_field_2_grid_all_levels(ds_new.z,lat_new2d,lon_new2d,lats_start,lons_start)
    u_interpolated = interpolate_field_2_grid_all_levels(ds_new.u,lat_new2d,lon_new2d,lats_start,lons_start)
    v_interpolated = interpolate_field_2_grid_all_levels(ds_new.v,lat_new2d,lon_new2d,lats_start,lons_start)
    w_interpolated = interpolate_field_2_grid_all_levels(ds_new.w,lat_new2d,lon_new2d,lats_start,lons_start)
    q_interpolated = interpolate_field_2_grid_all_levels(ds_new.q,lat_new2d,lon_new2d,lats_start,lons_start)
    r_interpolated = interpolate_field_2_grid_all_levels(ds_new.r,lat_new2d,lon_new2d,lats_start,lons_start)

    # Extract information at the level nearest to current pressure level

    z_interpolated = z_interpolated.isel(level = abs(p_new2d - p_aux_interpolated).argmin('level'))
    u_interpolated = u_interpolated.isel(level = abs(p_new2d - p_aux_interpolated).argmin('level'))
    v_interpolated = v_interpolated.isel(level = abs(p_new2d - p_aux_interpolated).argmin('level'))
    w_interpolated = w_interpolated.isel(level = abs(p_new2d - p_aux_interpolated).argmin('level'))
    q_interpolated = q_interpolated.isel(level = abs(p_new2d - p_aux_interpolated).argmin('level'))
    r_interpolated = r_interpolated.isel(level = abs(p_new2d - p_aux_interpolated).argmin('level'))
    
    # Storage

    z_traj = xr.concat([z_traj,z_interpolated],dim = 'time')
    u_traj = xr.concat([u_traj,u_interpolated],dim = 'time')
    v_traj = xr.concat([v_traj,v_interpolated],dim = 'time')
    w_traj = xr.concat([w_traj,w_interpolated],dim = 'time')
    q_traj = xr.concat([q_traj,q_interpolated],dim = 'time')
    r_traj = xr.concat([r_traj,r_interpolated],dim = 'time')
    try:
        p_new2d = p_new2d.drop('level', dim=None)
    except:
        pass
    p_traj = xr.concat([p_traj,p_new2d],dim = 'time')
    
    
    # 5. restart the loop
lat_traj = np.array(lat_traj)
lon_traj = np.array(lon_traj)
print(dt.datetime.now())
```
